# Regress√£o Linear Simples
##üç® Prevendo vendas de sorvetes com base na temperatura

Como primeira experi√™ncia aplicando uma regress√£o linear utilizando Python, vamos utilizar uma base de dados (Sorvetes.xlsx) para identificar se existe uma rela√ß√£o entre a temperatura em C com a venda de sorvetes.

```python
# Importando as bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


# Importando a base de dados
dados = pd.read_excel("Sorvete.xlsx")
dados.head()
```

Gr√°fico para melhor visualiza√ß√£o da rela√ß√£o vendas <-> temperatura.

```python
# Visualizando os dados
plt.scatter(dados['Temperatura'], dados['Vendas_Sorvetes'])
plt.xlabel('Temperatura (C)')
plt.ylabel('N√∫mero de vendas (milhares)')
plt.title('Rela√ß√£o entre Temperatura e Vendas de sorvetes')
plt.show()
```

√â poss√≠vel perceber que existe uma varia√ß√£o linear entre as duas vari√°veis, conforme uma cresce a outra tamb√©m cresce.
Agora vamos analisar a correla√ß√£o entre essas mesmas duas vari√°veis.

```python
dados.corr()
```

A correla√ß√£o entre as vari√°vies √© de 98.
Analisar a correla√ß√£o √© uma boa estrat√©gia para criar um modelo de regress√£o linear simples. A correla√ß√£o **mede a for√ßa e a dire√ß√£o do relacionamento linear entre as duas vari√°veis**, o que √© fundamental para modelos de regress√£o linear simples.

Ao realizar a an√°lise de correla√ß√£o √© poss√≠vel identificar quais **vari√°veis independentes (features) est√£o relacionadas com a vari√°vel dependente (resultado/previs√£o)**

Uma correla√ß√£o significativa (lembrando que a taxa vai de 0.0 at√© 1.0) indica que uma rela√ß√£o linear pode existir dentro dos dados e um modelo de regress√£o linear poderia ser aplicado.
